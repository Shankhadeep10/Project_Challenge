```python
df['published_month2']=df.published_month.apply(lambda x:calendar.month_abbr[int(x)])
df['published_month2'].columns=['value','month']
df['published_month2']
```

# Q13.Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig,(a1)=plt.subplots(1)
a1.scatter(df['view_count'],df['dislikes'],color="green")
a1.set_ylabel('dislikes')
a1.set_xlabel('view_count')
a1.set_title('scatter plot no.of view vs dislikes')
```

In this plot x axis column 'view_count' and y axis column 'dislike' both are positively correlated and have a same direction.

```python
correlation=df['view_count'].corr(df['dislikes'])
print(correlation)
if correlation > 0:
    print('There is a positive correlation between view and dislikes')
elif correlation < 0:
    print('There is a negative correlation between view and dislikes')
else:
    print("There is no significent correlation between view and dislikes")
```

On the above scatter plot I see that dislikes point is increase as well as numbers of view counts all increases and 
there is positive correlationship between view_count and dislikes.And the correlation value is 0.68 so we can say 68%
of people are disliked video.No its does not make any effects in this correlation.
